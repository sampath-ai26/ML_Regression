```python
import pandas as pd

dataset=pd.read_csv("50_Startups.csv")

dataset=pd.get_dummies(dataset, drop_first=True, dtype=int)
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent= dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
#from sklearn.preprocessing import StandardScaler   #section is added to check if r_score can be improved by doing STANDARDISATION
#sc=StandardScaler()
#x_train_scaled=sc.fit_transform(x_train)
#x_test_scaled=sc.transform(x_test)
x_train
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='linear', C=500)   #kernel = 'linear','rbf' - radial bais function for (NON-Linear)
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
