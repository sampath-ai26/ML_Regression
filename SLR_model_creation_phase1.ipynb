```python
import pandas as pd  #importing pandas library that has database related functions.
```

```python
dataset= pd.read_csv("salary_data.csv") #read_csv function is called to read data from csv source file and assigning the entire data to dataset variable
```

```python
dataset
```

```python
# spliting of I/P data from source and assigning into independent variable.
independent=dataset[["YearsExperience"]]
```

```python
# spliting of O/P data from source and assigning into dependent variable.
dependent=dataset[["Salary"]]
```

```python
# Datas available in independent and dependent variables are split up further into to training and test sets data with test_size 1/3
#since the train_test_split function returns 4 set values , hence we are assigning into 4 variables.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# This block creates the acutal model creation.
from sklearn.linear_model import LinearRegression
regressor=LinearRegression() #Model instance is created.
regressor.fit(x_train,y_train)   # model is trained and created ex y=wX+b i.e weight and bais values calculated completed after this step.
```

```python
# calculatd weight is stored in this default variable coef_
weight=regressor.coef_
```

```python
print(weight)
```

```python
# calculatd bais is stored in this default variable intercept_
bais=regressor.intercept_
print(bais)
```

```python
# For test set of data prediction is calculated here and stored in y_pred
y_pred=regressor.predict(x_test)
```

```python
# After model creation, using this we evaluate the created model efficacy to decide if created model is ok or not.
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
print(r_score) # not enough source data and hence getting low r2_score
```

```python
# Using this section we save the created model into filename.
import pickle
filename="finalized_model_liner.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# We load the saved model and execute and validate for the given input what is the predicted result.
loaded_model=pickle.load(open("finalized_model_liner.sav",'rb'))
result=loaded_model.predict([[10]])
```
