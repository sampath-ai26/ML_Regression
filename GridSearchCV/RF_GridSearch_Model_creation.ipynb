```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid={ 'criterion' :['squared_error','absolute_error','friedman_mse'],
             'max_features' : [None,'sqrt','log2'],
             'n_estimators' : [100]}

grid = GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
# grid will handle r_score grid values and training and test data split internally.
#Hence seperate x_train, x_Test, ,y_train, y _test not required
re=grid.cv_results_
print( " The R score value for the best params {}:" .format(grid.best_params_))
```

```python
# Evaluation will be handled by gridsearch in Cross validation
table = pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI : "))
children_input=int(input("Children:"))
sex_male_input=int(input("Sex Male 0 or1 :"))
smoker_yes_input=int(input("Smoker yes 0 or 1:"))
```

```python
future_prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future prediction {}" .format(future_prediction))
```
